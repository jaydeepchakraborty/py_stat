# Choose a number $U$ from the interval $[0,1]$ with uniform distribution

# Find the cumulative distribution and density for the random variables

## a) $Y = \left | U-\frac{1}{2} \right |$

## b) $Y = \left ( U-\frac{1}{2} \right )^{2}$

____

# a)

# We recall $F_{U}(u) = u$ and $f_{U}(u) = 1$

# $\phi(u) = \left | u-\tfrac{1}{2} \right | = \left\{\begin{matrix}\tfrac{1}{2} - u & ;u<\tfrac{1}{2}\\ u-\tfrac{1}{2} & ;u\geq\tfrac{1}{2}\end{matrix}\right.$

## Note: since $\phi(u)$ is not a monotonic function, we can't use the theorem from [Q2](02-Solution.ipynb)

# $\phi^{-1}(y) = \left\{\begin{matrix}\tfrac{1}{2} - y & ;y<\tfrac{1}{2}\\ y-\tfrac{1}{2} & ;y\geq\tfrac{1}{2}\end{matrix}\right.$

## To see how this function $\phi$ is working, let's take a random sample of 1,000 numbers in $[0,1]$ and see what values they map to

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
u = np.random.random(size = 1000)
plt.hist(u, bins = 10, normed = True)
plt.show()
```

### So our 100 numbers sampled from $[0,1]$ is pretty evenly spread across the possible values

```python
y = np.abs(u-0.5)
plt.hist(y, bins = 10, normed = True)
plt.show()
```

## The values are pretty evenly spread on the interval $[0,0.5]$

### This makes sense since no number between 0 and 1 can be further than 0.5 from 0.5, and the the sample is evenly distributed between 0 and 1 so the mapping should be evenly distributed among the distances from 0.5

# From this, we can assume that the pdf for $Y$ looks something like...

```python
x = np.arange(0,0.501,0.01)
y = 2*np.ones(len(x))
plt.plot(x,y)
plt.xlim([0,0.5])
plt.ylim([0,2.5])
plt.show()
```

# This function is given by $f_{Y}(y) = 2 \implies F_{Y}(y) = 2y$

____

# b)

## If we rename this $Y$ to be $Y^{*}$ and call the $Y$ from part a) $Y \implies Y^{*} = Y^{2}$

## The values of $Y$ are in $[0,1]$ and $\phi(y) = y^{2}$ which is monotonically increasing

## Therefore $F_{Y^{*}}(y^{*}) = F_{Y}(\phi^{-1}(y^{*}))$ 

## $\phi(y) = y^{2} \implies \phi^{-1}(y^{*}) = \sqrt{y^{*}} \implies F_{Y^{*}}(y^{*}) = 2\sqrt{y^{*}} \implies f_{Y^{*}}(y^{*}) = \frac{1}{\sqrt{y^{*}}}$
