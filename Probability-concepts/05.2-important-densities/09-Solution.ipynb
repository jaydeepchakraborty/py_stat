# Let $U$, $V$ be random numbers chosen independently from the interval $[0,1]$ with uniform distribution

# Find the cdf and pdf of the following


## a) $Y=U+V$

## b) $Y = \left | U-V \right |$

___

# a)

### Before we consider the continuous case, let's think about a similar discrete problem

### Let $U$ and $V$ represent the roll of a die

#### Then both $U$ and $V$ are uniformly distributed from $[1,6]$

## But the sum of their values are not uniformly distributed

### We can see this fairly easily by looking at all possible combinations that add up to 2 through 12

```python
results = []

for i in range(1,7):
    for j in range(1,7):
        print((i,j), i+j)
        results.append(i+j)
```

### We can plot this

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
plt.hist(results, bins = 100)
plt.xlim([0,14])
plt.show()
```

## We can see that rolling a total of 8 is more likely than a 2 or 12

# So we can predict a similar shape for *continuous* variables $U$ and $V$, with the edges at 0 and 2 with a peak at 1

```python
x = np.arange(0,2.01,0.01)
y = 1 - np.abs(x-1)
plt.plot(x,y)
plt.show()
```

# This function is equal to:

# $f(y) = \left\{\begin{matrix} y & y<1\\ 2-y & y\geq 1 \end{matrix}\right.$

# Which has a cdf of:

# $F(y) = \left\{\begin{matrix} \frac{1}{2}y^{2} & y<1\\ 2y-\frac{1}{2}y^{2} & y\geq 1 \end{matrix}\right.$

___

# b)

# Similarly to part a), we can look at the difference in the faces of two dice

```python
results = []

for i in range(1,7):
    for j in range(1,7):
        print((i,j), np.abs(i-j))
        results.append(np.abs(i-j))
```

```python
plt.hist(results, bins = 100)
plt.xlim([-1,6])
plt.show()
```

# We can see a similar shape for two 10-sided dice

```python
results = []

for i in range(1,11):
    for j in range(1,11):
        results.append(np.abs(i-j))
        
plt.hist(results, bins = 100)
plt.xlim([-1,10])
plt.show()
```

# What about 100-sided dice?

```python
results = []

for i in range(1,111):
    for j in range(1,111):
        results.append(np.abs(i-j))
        
plt.hist(results, bins = 1000)
plt.xlim([-1,110])
plt.show()
```

# We start to see that as the number of sides on the dice go higher, the line approaches a simple triangle shape

## The maximum value for $\left | U-V \right |$ is 1, and the minimum is 0

## So for *continuous* $U$ and $V$, we can assume the density function looks like:

```python
x = np.arange(0,1.01,0.01)
y = 2 - 2*x
plt.plot(x,y)
plt.show()
```

# This function is equal to:

# $f(y) = 2-2y$

# Which has a cdf of:

# $F(y) = 2y - y^{2}$
