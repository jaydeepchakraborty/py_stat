# 9. a)

# $Y=U+V$

# $f(y) = \left\{\begin{matrix} y & y<1\\ 2-y & y\geq 1 \end{matrix}\right.$

# $F(y) = \left\{\begin{matrix} \frac{1}{2}y^{2} & y<1\\ 2y-\frac{1}{2}y^{2} & y\geq 1 \end{matrix}\right.$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
n = 10000

U = np.random.random(size = n)
V = np.random.random(size = n)

Y = U+V

y1 = np.arange(0,1.01, 0.01)
y2 = np.arange(1,2.01, 0.01)
fy1 = y1
fy2 = 2-y2

y = np.append(y1, y2)
fy = np.append(fy1, fy2)

plt.hist(Y, bins = 100, normed = True)
plt.plot(y, fy, lw = 2, color = 'r')
plt.xlim([0,2])
plt.ylim([0,1])
plt.show()
```

# Looks good

# b)

# $Y = \left | U-V \right |$

# $f(y) = 2-2y$

# $F(y) = 2y - y^{2}$

```python
n = 10000

U = np.random.random(size = n)
V = np.random.random(size = n)

Y = np.abs(U-V)

y = np.arange(0,1.01, 0.01)
fy = 2-2*y

plt.hist(Y, bins = 100, normed = True)
plt.plot(y, fy, lw = 2, color = 'r')
plt.xlim([0,1])
plt.ylim([0,2])
plt.show()
```

# Looks good

# 10

# a)

# $Y = max \{ U,V \}$

# $f(y) = 2y$

# $F(y) = y^{2}$

```python
n = 10000

U = np.random.random(size = n)
V = np.random.random(size = n)

Y = np.maximum(U,V)

y = np.arange(0,1.01, 0.01)
fy = 2*y

plt.hist(Y, bins = 100, normed = True)
plt.plot(y, fy, lw = 2, color = 'r')
plt.xlim([0,1])
plt.ylim([0,2])
plt.show()
```

# Looks good

# b)

# $Y = min \{ U,V \}$

# $f(y) = 2-2y$

# $F(y) = 2y-y^{2}$

```python
n = 10000

U = np.random.random(size = n)
V = np.random.random(size = n)

Y = np.minimum(U,V)

y = np.arange(0,1.01, 0.01)
fy = 2-2*y

plt.hist(Y, bins = 100, normed = True)
plt.plot(y, fy, lw = 2, color = 'r')
plt.xlim([0,1])
plt.ylim([0,2])
plt.show()
```

# Looks good
