# A number $U$ is chosen at random in the interval $[0,1]$

# Find the probability that:

## a) $R=U^{2}<\frac{1}{4}$

## b) $S = U(1-U) < \frac{1}{4}$

## c) $T = \frac{U}{1-U} < \frac{1}{4}$

___

# a)


# First, let's create a plot to look at the possible values of R

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
U = np.arange(0,1.01,0.01)
R = U**2

plt.plot(U,R)
plt.axhline(0.25, color = 'r', ls = '--')
plt.axvline(0.5, color = 'r', ls = '--')
```

# We can see from the plot that $U<\frac{1}{2}\implies R < \frac{1}{4}$

# Well, $P(U<\frac{1}{2}) = \frac{1}{2} \implies P(R<\frac{1}{4}) = \frac{1}{2}$

___

# b)

# $ S = U(1-U)$

# Let's plot $S$

```python
U = np.arange(0,1.01,0.01)
S = U*(1-U)

plt.plot(U,S)
plt.show()
```

# We can see from the chart that the only value where $S\geq\frac{1}{4}$ is at $U=0.5$

## The probability of this $U$ value is negligible

# Therefore $P(S < \frac{1}{4}) = 1$

___

# c) $T = \frac{U}{1-U}$

```python
U = np.arange(0,1.01,0.01)
T = U/(1-U)

plt.plot(U,T)
plt.ylim([0,2])
plt.axhline(0.25, color = 'r', ls = '--')
plt.axvline(0.2, color = 'r', ls = '--')

plt.show()
```

# We can see from the plot that $U<0.2\implies T<\frac{1}{4}$

# $P(U<0.2) = 0.2 \implies P(T<\frac{1}{4}) = 0.2$
