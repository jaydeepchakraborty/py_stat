# Suppose we know a random variable $Y$ as a function of the uniform random variable $U: Y = \phi(U)$ 

# Suppose we have calculated the cumulative distribution function $F_{Y}(y)$ and also the density function $f_{Y}(y)$

# We can check the accuracy of our calculation by taking random values of $U$ and plotting a histogram of $\phi(rnd)$

# If the plot looks similar to that of  $f_{Y}(y)$, then we did it right

# Check the results of [Q1](01-Solution.ipynb) and [Q2](02-Solution.ipynb) using this method

___

# 1. a)

# $U$ is from the unit interval $[0,1]$

# $\phi(u) = u+2$

# $f_{Y}(y)  = 1$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
rand = np.random.random(size = 10000)
u = np.arange(0,1.01,0.01)
y = u+2
fy = np.ones(len(y))
phi_u = rand+2
plt.plot(y,fy)
plt.hist(phi_u, normed = True)
```

___

# 1. b)

# $f_{Y}(y) = \frac{1}{3}y^{\frac{-2}{3}}$

```python
rand = np.random.random(size = 10000)
phi_u = rand**3

u = np.arange(0.01,1.01,0.01)
y = u**3
fy = 0.333*((y)**(-0.66666))
plt.plot(y,fy)
plt.hist(phi_u, bins = 100, normed = True)
plt.ylim([0,10])
```

___

# 2. a)

# $\phi(u) = \frac{1}{u+1}$

# $f_{Y}(y) =  \frac{1}{y^{2}}$

```python
rand = np.random.random(size = 10000)
phi_u = 1/(rand+1)

u = np.arange(0.01,1.01,0.01)
y = 1/(u+1)
fy = 1/(y**2)
plt.plot(y,fy)
plt.hist(phi_u, bins = 100, normed = True)
plt.show()
```

___

# 2. b)

# $\phi(u) = log(u+1)$

# $f_{Y}(y) = e^{y}$

```python
rand = np.random.random(size = 10000)
phi_u = np.log(rand+1)

u = np.arange(0.00,1.01,0.01)
y = np.log(u+1)
fy = np.exp(y)
plt.plot(y,fy)
plt.hist(phi_u, bins = 100, normed = True)
plt.show()
```
