## We kinda cheated and used this technique to derive $f_{Y}(y)$ but anyways...

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
rand = np.random.random(size = 1000)
phi = np.abs(rand-0.5)

u = np.arange(0,1.01, 0.01)
y = np.abs(u-0.5)
f_y = 2*np.ones(len(y))

plt.plot(y,f_y)
plt.hist(phi, bins = 10, normed = True)
plt.show()
```

```python
rand = np.random.random(size = 1000)
phi = (rand-0.5)**2

u = np.arange(0,1.01, 0.01)
y = (u-0.5)**2
f_y = 1/np.sqrt(y+0.0000001)

plt.plot(y,f_y)
plt.hist(phi, bins = 10, normed = True)
plt.ylim([0,10])
plt.show()
```
