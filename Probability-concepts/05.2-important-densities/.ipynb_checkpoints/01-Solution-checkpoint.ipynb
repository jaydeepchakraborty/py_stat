# Choose a number $U$ from the unit interval $[0,1]$ with uniform distribution

# Find the cumulative distribution and density for the random variables

## a) $Y = U+2$

## b) $Y = U^{3}$

___

# a)

# $F_{U}(u) = P(U\leq u) = u$ for $u$ in $[0,1]$

# $f_{U}(u)  = \frac{d}{du}F_{U}(u)= 1$

# $F_{Y}(y) = P(Y\leq y) = P(U+2 \leq y) = y-2$ for $y$ in $[2,3]$ 

# $f_{Y}(y)  = \frac{d}{dy}F_{Y}(y)= 1$

___

# b)

# First, let's look at the plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
x = np.arange(0,1,0.01)
y = x**3
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('P(Y<y)')
plt.xlabel('u')
plt.plot(x,y)
```

# $F_{Y}(y) = P(Y\leq y) = P(U^{3}\leq y) =P(U\leq y^{\frac{1}{3}}) = y^{\frac{1}{3}} \implies f_{Y}(y) = \frac{1}{3}y^{\frac{-2}{3}}$
