# Let $U$ and $V$ be random numbers chosen independently from the interval $[0,1]$ 

# Find the cdf and pdf for the following:

## a) $Y = max \{ U,V \}$

## b) $Y = min \{ U,V \}$

____

# a)

# Like in [Q9](09-Solution.ipynb), let's start by thinking about two dice

```python
results = []

for i in range(1,7):
    for j in range(1,7):
        print((i,j),max(i,j))
        results.append(max(i,j))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
plt.hist(results, bins = 100)
plt.xlim([0,7])
plt.show()
```

# If the dice each have 100 sides:

```python
results = []

for i in range(1,101):
    for j in range(1,101):
        results.append(max(i,j))
        
plt.hist(results, bins = 1000)
plt.xlim([0,101])
plt.show()
```

# Seems to be a simple linear function


# Therefore the pdf of $Y$ is equal to:

# $f(y) = 2y$

# And the cdf is equal to:

# $F(y) = y^{2}$

___

# b)

```python
results = []

for i in range(1,7):
    for j in range(1,7):
        results.append(min(i,j))
        
plt.hist(results, bins = 100)
plt.xlim([0,7])
plt.show()
```

```python
results = []

for i in range(1,101):
    for j in range(1,101):
        results.append(min(i,j))
        
plt.hist(results, bins = 100)
plt.xlim([0,101])
plt.show()
```

# Same as a), but in the opposite direction

# $f(y) = 2-2y$

# $F(y) = 2y-y^{2}$
