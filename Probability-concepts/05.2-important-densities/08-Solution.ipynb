## We can take a sample by taking 1,000 random values from $[0,1]$ then taking $\sqrt{rnd}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
rand = np.random.random(size = 10000)
phi = np.sqrt(rand)

y_hist, x_hist = np.histogram(phi, bins = 100)

x = np.arange(0.00,1.01,0.01)
y = x**2

plt.plot(x,y)
plt.plot(x_hist[:-1], np.true_divide(np.cumsum(y_hist),np.sum(y_hist)))
plt.show()
```
