# A large number of waiting time problems have an *exponential distribution* of outcomes

# We shall see (in Section 5.2) that such outcomes are simulated by computing $\frac{-1}{\lambda}log(rnd)$, where $\lambda > 0$

# For waiting times produced this way, the average waiting time is $\frac{1}{\lambda}$

## For example, the times spent waiting for a car to pass on a highway, or the times between emissions of particles from a radioactive source, are simulated by a sequence of random numbers, each of which is computing $\frac{-1}{\lambda}log(rnd)$ where $\frac{1}{\lambda}$ is the average time between cars or emissions

# Write a program to simulate the times between cars when the average time between them is 30 seconds

# Have your program plot an area bar graph for these times by breaking the time interval from 0 to 120 into 24 subintervals

# On the same pair of axes, plot the function $f(x) = \frac{1}{30}e^{-\frac{1}{30}x}$. Does the function fit the graph well?

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
def sim(n):
    rand_array = np.random.random(size = n)
    results = np.log(rand_array)*(-30)
    return(results)
```

```python
counts, bins = np.histogram(sim(10000), bins = 24, range=(0,120), normed = True)
```

```python
plt.plot(bins[:-1], counts)
constant = float(1)/30
plt.plot(bins[:-1], constant*np.exp(-constant*bins[:-1]), ls = 'dashed', color = 'r')
```

# Seems to fit pretty closely
