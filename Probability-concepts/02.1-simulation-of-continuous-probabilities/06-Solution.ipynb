# To simulate the Buffon's needle problem we choose independently the distance $d$ and the angle $\theta$ at random, with $0\leq d \leq \frac{1}{2}$ and $0\leq \theta \leq \frac{\pi}{2}$ and test whether $d \leq (\frac{1}{2})sin\theta$

# Doing this a large number of times, we estimate $\pi$ as $\frac{2}{a}$, where $a$ is the proportion of times the condition is satisfied

## Before we run the simulation, let's think about this for a second

### You have a surface with parallel lines that are distance $d$ apart

![](images/6-1.png)

### You drop a needle of length $l$ onto the surface

### Now, we know that if the distance from the center of the needle to the closest line is $x$, and the acute angle formed by the needle is $\theta$...

![](images/6-2.png)

### For simplicity, let's assume $l=1$ and $d=1$

### Then we know that if $x>\frac{1}{2}$, then the needle won't reach the line so the probability of crossing it is ZERO

### But since the distance between the lines is 1, the $x$ value can never be greater than $\frac{1}{2}$, i.e. the needle is aways in reach

### So, the needle crosses the line when $\frac{1}{2}sin\theta \geq x$

### $\theta$ can have a value in the interval $[0, \pi]$ i.e. between 0 and 180 degrees

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
x = np.arange(0, np.pi, step = 0.1)
y = 0.5*np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('x')
plt.axhline(0.5, ls = 'dashed', color = 'r')
```

# We can use the graph above to answer the following question: for fixed $\theta$, what are the possible values for $x$ where the needle will cross the line?

## For each angle on the x-axis, the area below the blue curve represents $x$ values where $x\leq sin(\theta)$

## Therefore, if we take the ratio of the area under the curve to the total area, we'll get the proportion of $x$ values that satisfy the condition

```python
plt.plot(x,y)
plt.axvline(np.pi*0.25, ls = 'dashed', color = 'g')
plt.axhline(0.5, ls = 'dashed', color = 'r')
plt.axhline(0, ls = 'dashed', color = 'r')
plt.fill_between(x, y, 0, facecolor='b')
plt.fill_between(x, 0.5, y, facecolor='r')
plt.ylim([-0.01, 0.51])
```

# The blue area on the chart can be solved by integrating $y = 0.5sin(\theta)$

# $\int^{\pi}_{0}0.5\cdot sin(\theta)d\theta = 0.5\cdot (-cos(\pi) - (-cos(0)) = 1$

# The sum of the blue area and the red area is simply the area of the rectangle given by $A=0.5\cdot \pi = \frac{\pi}{2}$

# So the probability is equal to $P(\text{Crosses line}) = \frac{1}{\frac{\pi}{2}} = \frac{2}{\pi}$

___
# So what does this mean for the simulation?

### If we generate random $x$ and random $\theta$ and test to see if $x<0.5sin\theta$, the proportion of trials that cross the line should be approximately equal to $\frac{2}{\pi}$

```python
def check(x, theta):
    return 1*(x<np.sin(theta)*0.5)
```

```python
check = np.vectorize(check)
```

```python
def Buffon_sim(n):
    x_array = 0.5*np.random.random(size = n)
    theta_array = np.pi*0.5*np.random.random(size = n)
    return(np.sum(check(x_array, theta_array))/float(n))
```

```python
Buffon_sim(1000)
```

```python
for n in [100, 1000, 10000, 100000, 1000000]:
    print('n = '+str(n))
    print('pi estimate = '+str(2/Buffon_sim(n)))
    print('\n')
```
