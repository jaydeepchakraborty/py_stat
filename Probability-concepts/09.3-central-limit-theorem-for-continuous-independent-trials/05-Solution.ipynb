# Write a program to choose independently 25 numbers at random from $[0,20]$, compute their sum, and repeat this experiment 1000 times

# Make a bar graph and compare it to the normal approximation

# Do the same for $S_{25}^{*}$ and $A_{25}$

_______

# a) $S_{25}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from scipy.stats import norm
```

```python
n = 25
n_trials = 1000

random_array = np.random.uniform(0, 20, size = [n, n_trials])
sum_array = np.sum(random_array, axis = 0)

mu = n*10
sigma = np.sqrt(n*400.0/12)

x = np.arange(mu - 3*sigma, mu + 3*sigma)
norm_array = norm.pdf(x, loc = mu, scale = sigma)

plt.plot(x, norm_array)
plt.hist(sum_array, normed = True, bins = 30)
plt.show()
```

______

# b) $S_{25}^{*}$

```python
standardized_array = (sum_array - mu)/(sigma)
x = np.arange(-3, 3, 0.01)
norm_array = norm.pdf(x)

plt.plot(x, norm_array)
plt.hist(standardized_array, normed = True, bins = 30)
plt.show()
```

______

# c) $A_{n}$

```python
random_array = np.random.uniform(0, 20, size = [n, n_trials])
sum_array = np.sum(random_array, axis = 0)

mean_array = sum_array/25

mu = 10
sigma = np.sqrt(400.0/(12*25))

x = np.arange(mu - 3*sigma, mu + 3*sigma, 0.1)

norm_array = norm.pdf(x, loc = mu, scale = sigma)

plt.plot(x, norm_array)
plt.hist(mean_array, normed = True, bins = 30)
plt.show()
```
