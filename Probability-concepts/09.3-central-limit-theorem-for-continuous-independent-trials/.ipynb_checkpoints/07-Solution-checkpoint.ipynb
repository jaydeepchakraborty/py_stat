# The CLT says the sums of independent random variables tend to look normal, no matter the distribution

# Let us test this by a computer simulation

# Choose independently 25 numbers from the interval $[0,1]$ with probability density defined below and compute the sum $S_{25}$

# Repeat this experiment 1000 times and make a bar graph of the results

# Now plot $\phi(x)$

# How well does the normal distribution fit the plot?

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from scipy.stats import norm
```

_____

# a)

# $f(x) = 1$

# This is simply the uniform distribution on $[0,1]$

```python
random_array = np.random.random(size = [25, 1000])
sum_array = np.sum(random_array, axis = 0)
```

```python
mu = 0.5*25
sigma = np.sqrt(25.0/12)
```

```python
x = np.arange(mu - 3*sigma, mu+3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)

plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```

_______

# b)

# $f(x) = 2x \implies F(X) = \int 2x dx = x^{2} \implies F^{-1}(x) = \sqrt{x}$

# Therefore, to create the random samples from the distribution, we create an array of random numbers in $[0,1]$ and see where it lies on the cdf

```python
random_array = np.random.random(size = [25, 1000])
mapped_array = np.sqrt(random_array)
sum_array = np.sum(mapped_array, axis = 0)
```

# $E(X) = \int_{0}^{1}x\cdot 2x dx = \frac{2}{3}$ 

# $V(X) = \int_{0}^{1}2x(x-\frac{2}{3})^{2} dx = \int_{0}^{1}2x\left ( x^{2} - \frac{4x}{3} + \frac{4}{9}\right ) dx$
# $ = \int_{0}^{1}\left ( 2x^{3} - \frac{8x^{2}}{3} + \frac{8x}{9} \right ) dx = \frac{1}{2} - \frac{8}{9} + \frac{4}{9} = \frac{1}{18}$

# $\implies E(S_{25}) = \frac{50}{3}$ and $V(S_{25}) = \frac{25}{18}$

```python
mu = 50.0/3
sigma = 5/np.sqrt(18)
```

```python
x = np.arange(mu - 3*sigma, mu + 3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)
```

```python
plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```

# c)

# $f(x) = 3x^{2} \implies F(x) = x^{3} \implies F^{-1}(x) = x^{\frac{1}{3}}$

```python
random_array = np.random.random(size = [25, 1000])
mapped_array = np.power(random_array, 1.0/3)
sum_array = np.sum(mapped_array, axis = 0)
```

# $E(X) = \int_{0}^{1}3x^{2}\cdot x dx = \frac{3}{4}$

# $E(X^{2}) =  \int_{0}^{1}3x^{2}\cdot x^{2} dx = \frac{3}{5}$

# $\implies V(X) = \frac{3}{5} - \frac{9}{16} = \frac{3}{80}$

# $\implies E(S_{25}) = \frac{75}{4}$ and $V(S_{25}) = \frac{75}{80}$

```python
mu = 75.0/4
sigma = np.sqrt(75.0/80)
```

```python
x = np.arange(mu - 3*sigma, mu + 3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)
```

```python
plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```

# d)

# $f(x) = 4 \left | x - \frac{1}{2} \right | = \left\{\begin{matrix}4 \left ( \frac{1}{2} - x \right )& x \leq \frac{1}{2} \\ 4 \left ( x - \frac{1}{2} \right )& x > \frac{1}{2}\end{matrix}\right.$

# $\implies F(X) =  \left\{\begin{matrix}2x - 2x^{2} & x\leq \frac{1}{2}\\1 + 2x^{2} - 2x & x > \frac{1}{2}\end{matrix}\right.$

```python
x1 = np.arange(0,0.5,0.001)
x2 = np.arange(0.5, 1, 0.001)
f_x1 = 2 - 4*x1
f_x2 = 4*x2 - 2

F_x1 = 2*x1 - 2*(x1**2)
F_x2 = 1 + 2*(x2**2) - 2*x2

plt.plot(x1, f_x1, color = 'r', lw = 3)
plt.plot(x2, f_x2, color = 'r', lw = 3, label = 'f(x)')

plt.plot(x1, F_x1, color = 'g', lw = 3)
plt.plot(x2, F_x2, color = 'g', lw = 3, label = 'F(x)')
```

# $\implies F^{-1}(x) = \left\{\begin{matrix}\frac{1}{2}(1 - \sqrt{1 - 2x}) & x \leq \frac{1}{2}\\ \frac{1}{2}(\sqrt{2x-1} + 1) & x > \frac{1}{2}\end{matrix}\right.$

```python
def F_inv(x):
    if x <= 0.5:
        val = 0.5*(1 - np.sqrt(1 - 2*x))
    else:
        val = 0.5*(np.sqrt(2*x - 1) + 1)
    return val
```

```python
F_inv = np.vectorize(F_inv)
```

```python
random_array = np.random.random(size = [25, 1000])
mapped_array = F_inv(random_array)
sum_array = np.sum(mapped_array, axis = 0)
```

# $E(X) = \int_{0}^{\frac{1}{2}}2x - 4x^{2} dx + \int_{\frac{1}{2}}^{1}4x^{2}-2xdx = \frac{1}{2}$

# $E(X^{2}) = \int_{0}^{\frac{1}{2}}2x^{2} - 4x^{3} dx + \int_{\frac{1}{2}}^{1}4x^{3}-2x^{2}dx = \frac{3}{8}$

# $V(X) = \frac{3}{8} - \frac{1}{4} = \frac{1}{8}$

# $\implies E(S_{25}) = \frac{25}{2}$ and $V(X) = \frac{25}{8}$

```python
mu = 25.0/2
sigma = 5.0/np.sqrt(8)
```

```python
x = np.arange(mu - 3*sigma, mu + 3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)
```

```python
plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```

# e)

# $f(x) = 2 - 4\left | x - \frac{1}{2}\right | = \left\{\begin{matrix}2 - 4 \left ( \frac{1}{2} - x \right )& x \leq \frac{1}{2} \\2 -  4 \left ( x - \frac{1}{2} \right )& x > \frac{1}{2}\end{matrix}\right. $

# $= \left\{\begin{matrix}4x& x \leq \frac{1}{2} \\ 4 - 4x& x > \frac{1}{2}\end{matrix}\right.$

# $\implies F(x) = \left\{\begin{matrix}2x^{2} & x \leq \frac{1}{2}\\ 4x - 2x^{2} - 1 & x > \frac{1}{2}\end{matrix}\right.$

```python
x1 = np.arange(0,0.5,0.001)
x2 = np.arange(0.5, 1, 0.001)
f_x1 = 4*x1
f_x2 = 4 - 4*x2

F_x1 = 2*(x1**2)
F_x2 = 4*x2 - 2*(x2**2) - 1

plt.plot(x1, f_x1, color = 'r', lw = 3)
plt.plot(x2, f_x2, color = 'r', lw = 3, label = 'f(x)')

plt.plot(x1, F_x1, color = 'g', lw = 3)
plt.plot(x2, F_x2, color = 'g', lw = 3, label = 'F(x)')
```

# $\implies F^{-1}(x) = \left\{\begin{matrix}\sqrt{\frac{x}{2}} & x \leq \frac{1}{2}\\ 1 - \frac{\sqrt{1 - x}}{\sqrt{2}} & x > \frac{1}{2}\end{matrix}\right.$

```python
def F_inv(x):
    if x<= 0.5:
        val = np.sqrt(x/2.0)
    else:
        val = 1 - np.sqrt(1-x)/np.sqrt(2)
    return val
```

```python
F_inv = np.vectorize(F_inv)
```

```python
random_array = np.random.random(size = [25, 1000])
mapped_array = F_inv(random_array)
sum_array = np.sum(mapped_array, axis = 0)
```

# $E(X) = \int_{0}^{\frac{1}{2}}4x\cdot xdx + \int_{\frac{1}{2}}^{1}\left (4 - 4x \right )xdx = \frac{1}{6} + \frac{1}{3} = \frac{1}{2}$
# $E(X^{2}) = \int_{0}^{\frac{1}{2}}4x^{3}dx + \int_{\frac{1}{2}}^{1}4x^{2} - 4x^{3}dx = \frac{1}{16} + \frac{11}{48} = \frac{7}{24}$

# $V(X) = \frac{7}{24} - \frac{1}{4} = \frac{1}{24}$

# $\implies E(S_{25}) = \frac{25}{2}$ and $V(S_{25}) = \frac{25}{24}$

```python
mu = 25.0/2
sigma = np.sqrt(25.0/24)
```

```python
x = np.arange(mu - 3*sigma, mu + 3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)
```

```python
plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```
