# Repeat the experiment described in [Exercise 7](07-Solution.ipynb) but now choose 25 numbers from $[0,\infty)$ using $f(x) = e^{-x}$

________

# $f(x) = e^{-x} \implies F(X) = 1 - e^{-x} \implies F^{-1}(x) = -ln(1-x)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from scipy.stats import norm
```

```python
def F_inv(x):
    return -np.log(1-x)
```

```python
F_inv = np.vectorize(F_inv)
```

```python
random_array = np.random.random(size = [25, 1000])
mapped_array = F_inv(random_array)
sum_array = np.sum(mapped_array, axis = 0)
```

# $E(X) = \int_{0}^{\infty}xe^{-x}dx = 1$

# $E(X^{2}) =  \int_{0}^{\infty}x^{2}e^{-x}dx = 2$

# $V(X) = E(X^{2}) - (E(X))^{2} = 2 - 1 = 1$

# $\implies E(S_{25}) = 25$ and $V(S_{25}) = 25$

```python
mu = 25
sigma = 5
```

```python
x = np.arange(mu - 3*sigma, mu+3*sigma, 0.1)
norm_array = norm.pdf(x, loc = mu, scale = sigma)
```

```python
plt.hist(sum_array, bins = 30, normed = True)
plt.plot(x, norm_array, lw = 4)
plt.show()
```
