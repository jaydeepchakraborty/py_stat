# In general, the CLT gives a better estimate than Chebyshev's inequality for the average of a sum

# To see this, take $A_{25}$ and let $N$ be its normal approximation

# Provide a table of the function $F(x) = P(|A_{25} - 10|\geq x)$

# $= \left ( \text{Fraction of the total of 1000 trials for which } |A_{25}| \geq x \right )$

# Do the same for $N(x)$

# Plot these two functions, along with the Chebyshev function $g(x) = \frac{4}{(3x^{2})}$

___________

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from scipy.stats import norm
```

```python
mu = 10
sigma = np.sqrt(400.0/12)/5
```

```python
random_array = np.random.uniform(0, 20, size = [25, 1000])
mean_array = np.mean(random_array, axis = 0)
```

```python
def F(x):
    diff_array = np.abs(mean_array - 10)
    prop = np.mean(diff_array >= x)
    return prop
```

```python
def N(x):
    z = x/sigma
    return 2*norm.cdf(-z)
```

```python
def g(x):
    return 4/(3*((x+0.000000000001)**2))
```

```python
F = np.vectorize(F)
N = np.vectorize(N)
g = np.vectorize(g)
```

```python
x = np.arange(0, 20, 0.5)

F_array = F(x)
N_array = N(x)

g_array = np.minimum(g(x), 1)
```

```python
plt.plot(x, F_array, label = 'F(x)')
plt.plot(x, N_array, label = 'N(x)')
plt.plot(x, g_array, label = 'g(x)')
plt.legend()
plt.xlim(0,10)
plt.show()
```
