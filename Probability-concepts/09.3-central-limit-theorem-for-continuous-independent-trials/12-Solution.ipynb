# First, choose 364 numbers $X_{i}$ with density $f(x) = N(x, 0, 0.25)$

# Now, form the sum $Y_{365} = 100 + X_{1} + X_{2}+...+X_{364}$, and repeat the experiment 200 times

# Make a histogram on $[50, 150]$ and superimpose the normal density

_______________

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from scipy.stats import norm
```

```python
random_array = np.random.normal(loc = 0, scale = 0.25, size = [364, 200])
sum_array = 100 + np.sum(random_array, axis = 0)
```

```python
x = np.arange(50, 150, 0.5)
norm_array = norm.pdf(x, loc = 100, scale = np.sqrt(365.0/4))
```

```python
plt.hist(sum_array, bins = 60, normed = True)
plt.plot(x, norm_array, lw = 3)
```
