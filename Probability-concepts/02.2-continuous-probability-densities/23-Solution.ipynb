# Write a program that picks a random number between 0 and 1 and computes the negative of its logarithm

# Repeat this process a large number of times and plot the histogram

# On this chart, also plot the density $f(x) = e^{-x}$

# How well does this density fit?

```python
import numpy as np
```

```python
def sim(n):
    rand = np.random.random(size = n)
    return -np.log(rand)
```

```python
results = sim(10000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
plt.hist(results, bins = 100, normed = True)

x = np.arange(0, 10.01, 0.01)
y = np.exp(-x)

plt.plot(x,y)
plt.show()
```

# Fits great!
