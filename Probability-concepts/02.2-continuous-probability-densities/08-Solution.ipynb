# Choose independently two numbers $B$ and $C$ *at random* from the interval [0,1] with uniform density

# Note that the point $(B,C)$ is then chosen *at random* in the unit square.

# Find the probability that:

### a) $B+C<1/2$

### b) $BC < \frac{1}{2}$

### c) $\left | B-C \right | < \frac{1}{2}$

### d) $max \left \{ B, C \right \} < \frac{1}{2}$

### e) $min \left \{ B, C \right \} < \frac{1}{2}$

### f) $B<\frac{1}{2} \text{ AND } (1-C) < \frac{1}{2}$

### g) Conditions c) and f) both hold

### h) $B^{2} + C^{2} < \frac{1}{2}$

### i) $\left ( B - \frac{1}{2} \right )^{2} + \left ( C - \frac{1}{2} \right )^{2} < \frac{1}{4}$

___
# a)

### Let's visualize this condition

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
x = np.arange(0,0.5,0.01)
y = 0.5 - x
plt.fill_between(x,0,y)
plt.axhline(1, ls = 'dashed', color = 'r')
plt.axvline(1, ls = 'dashed', color = 'r')
plt.xlim(-0.01, 1.01)
plt.ylim(-0.01, 1.01)
```

### The blue area represents the combinations of $B$ and $C$ that satisfy the condition

### $(\text{Area of blue triangle}) = \frac{0.5\cdot 0.5}{2} = \frac{1}{8}$

### $(\text{Total Area}) = 1\cdot1 = 1$

## So, $P(B+C<\frac{1}{2}) = \frac{(\text{Area of blue triangle})}{(\text{Total Area})} = \frac{1}{8}$

___
# b)

### Let's look at the plot

```python
x = np.arange(0,1,0.01)
y = 1/(2*x)
plt.fill_between(x,0,y)
plt.fill_between(x,1,y, color = 'r')
plt.axhline(0.5, ls = 'dashed', color = 'y')
plt.axvline(0.5, ls = 'dashed', color = 'y')
plt.ylim([0,1])
```

### The blue area represents the combinations of $B$ and $C$ whose product is less than 0.5

### The red area starts after 0.5, so we can calculate the red area as 0.5 minus the area under the curve $f(x) = 0.5/x$ from 0.5 to 1

# $\int^{1}_{0.5}\frac{0.5}{x}dx = 0.5\cdot \int^{1}_{0.5}\frac{1}{x}dx = 0.5\cdot \left ( ln(x)|^{1}_{0.5}\right )$

# $= 0.5 (ln(1) - ln(0.5)) = 0.5(0.693) = 0.3466$

# So, $(\text{Red Area}) = 0.5 - 0.3466 = 0.1534$

# Therefore,

# $P(BC < \frac{1}{2}) = \frac{(\text{Blue Area})}{(\text{Total Area})} = \frac{1-(\text{Red Area})}{1} = 0.8466$

___
# c)

### Looking at the plot

```python
x1 = np.arange(0,0.51,0.01)
x2 = np.arange(0.5,1,0.01)
y1 = (0.5 + x1)
y2 = x2 - 0.5
plt.fill_between(x1,0,y1, color = 'b')
plt.fill_between(x2,y2,1, color = 'b')
plt.fill_between(x1,1,y1, color = 'r')
plt.fill_between(x2,0,y2, color = 'r')
```

### We can just calculate the areas of the red triangles and taking 1 minus their combined area gives the blue area

## $(\text{Area of red triangles}) = 2 \cdot \frac{0.5\cdot0.5}{2} = \frac{1}{4}$

# So, $(\text{Blue Area}) = 1 - \frac{1}{4} = \frac{3}{4} \implies P(\left | B-C \right | < \frac{1}{2}) = \frac{3}{4}$

___
# d)

### Looking at the plot

```python
x = np.arange(0,0.5,0.01)
plt.fill_between(x, 0.5, 0)
plt.fill_between(x, 1, 0.5, color = 'r')
plt.fill_between(np.arange(0.49, 1,0.01), 1, 0, color = 'r')
```

### The blue area represents the only possible combinations, so we just need to calculate its area

# $P(max \left \{ B, C \right \} < \frac{1}{2}) = 0.5\cdot 0.5 = \frac{1}{4}$

___
# e)

### Looking at the plot

```python
plt.fill_between(np.arange(0,0.5,0.01), 1, 0, color = 'b')
plt.fill_between(np.arange(0.49,1,0.01), 0.5, 0, color = 'b')
plt.fill_between(np.arange(0.49,1,0.01), 1, 0.5, color = 'r')
```

# $P(min \left \{ B, C \right \} < \frac{1}{2}) = \frac{3}{4}$

___
# f)

# $ (1-C) < \frac{1}{2} \implies C > 1/2$

### So now, let's look at the plot

```python
plt.fill_between(np.arange(0,0.5,0.01), 1, 0.5, color = 'b')
plt.fill_between(np.arange(0,0.5,0.01), 0.5, 0, color = 'r')
plt.fill_between(np.arange(0.49,1,0.01), 1, 0, color = 'r')
```

# $P(B<\frac{1}{2} \text{ AND } (1-C) < \frac{1}{2}) = \frac{1}{4}$

___
# g)

### We can look at the two plots to see where they overlap

```python
x1 = np.arange(0,0.51,0.01)
x2 = np.arange(0.5,1,0.01)
y1 = (0.5 + x1)
y2 = x2 - 0.5
plt.fill_between(x1,0,y1, color = 'b', alpha = 0.5)
plt.fill_between(x2,y2,1, color = 'b', alpha = 0.5)
plt.fill_between(x1,1,y1, color = 'r', alpha = 0.5)
plt.fill_between(x2,0,y2, color = 'r', alpha = 0.5)

plt.fill_between(np.arange(0,0.5,0.01), 1, 0.5, color = 'b', alpha = 0.5)
plt.fill_between(np.arange(0,0.5,0.01), 0.5, 0, color = 'r', alpha = 0.5)
plt.fill_between(np.arange(0.49,1,0.01), 1, 0, color = 'r', alpha = 0.5)
```

### The only overlapping blue areas lie in the top left

# $P(\text{c. and f. both hold}) = 0.5 \cdot \frac{1}{4} = \frac{1}{8}$

___
# h)

### Looking at the plot

```python
x = np.arange(0,1,0.01)
y = np.nan_to_num(np.sqrt(0.5 - x**2))
plt.fill_between(x,0,y, color = 'b')
plt.fill_between(x,1,y,color = 'r')
plt.xlim([0,1])
```

### To solve for the blue area, we could integrate the area under the function $f(x) = \sqrt{\frac{1}{2} - x^{2}}$ from 0 to $\frac{1}{\sqrt{2}}$

# THIS IS A HUGE PAIN IN THE ASS

### Instead, we know that the blue area forms one quarter of a circle with radius $\frac{1}{\sqrt{2}}$, therefore its area is:

## $ (\text{Area of blue region}) = \pi (\frac{1}{\sqrt{2}})^{2} \cdot \frac{1}{4} = \frac{\pi}{8}$

### So $P(B^{2} + C^{2} < \frac{1}{2}) = \frac{\pi}{8}$

# i)

### Let's look at the plot

```python
x = np.arange(0,1,0.01)
y1 = 0.5 + np.sqrt(0.25 - (x - 0.5)**2)
y2 = 0.5 - np.sqrt(0.25 - (x - 0.5)**2)
```

```python
plt.fill_between(x,y1,y2, color = 'b')
plt.fill_between(x,1,y1, color = 'r')
plt.fill_between(x,0,y2, color = 'r')
```

### This condition simply means the squared distance from the centre of the circle is less than $\frac{1}{4}$ i.e. within $\frac{1}{2}$ of the centre

### So $P(\left ( B - \frac{1}{2} \right )^{2} + \left ( C - \frac{1}{2} \right )^{2} < \frac{1}{4}) = \frac{\pi}{4}$
