# Suppose you choose *at random* a real number $X$ from the interval $[2,10]$

## a) Find the density function $f(x)$ and the probability of an event $E$ for this experiment, where $E$ is a subinterval $[a,b]$ of $[2,10]$

## b) From a), find the probability that:

### i) $P(5<X)$
### ii) $P(5<X<7)$
### iii) $P(X^{2}-12X+35>0)$

___
# a)

## We know $\int^{b}_{a}f(x)dx = 1$ and we know that the distribution is uniform since we draw the number at random

## Therefore $F(X<x) = \frac{x-a}{10-2} = \frac{x-a}{8}$

## So $f(x) = \frac{d}{dx}F(X<x) = \frac{d}{dx}\frac{x-a}{8} = \frac{1}{8}$

# b)

## i)

# $P(5<X) = \int^{5}_{10}\frac{1}{8} = \frac{10-5}{8} = \frac{5}{8}$

## ii)

# $P(5<X<7) = \int^{5}_{7}\frac{1}{8} = \frac{7-5}{8} = \frac{1}{4}$

## iii)

## First, let's look at the function $X^{2} - 12X + 35$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
x = np.arange(2, 10, 0.1)
y = x**2 -12*x + 35
```

```python
plt.plot(x,y)
plt.axhline(0, ls = 'dashed', color = 'r')
```

# So the condition is satisfied for values above the red line (shown below as the area in the plot below)

```python
plt.plot(x,y)
plt.axhline(0, ls = 'dashed', color = 'r')
plt.fill_between(x, y*(y>0), 0, color = 'g')
```

# So we want the area under the curve for values less than 5, or greater than 7

# We could've also derived this as $X^{2} - 12X + 35 = (X-5)(X-7)$

# $P(X<5)+P(X>7) = 1 - P(5<X<7)  = 1 - \frac{1}{4} = \frac{3}{4}$
