# Take a stick of unit length and break it into 3 pieces, choosing the break points at random

# What is the probability that the three pieces can be used to form a triangle?

# *Hint: the sum of the lengths of any two pieces must exceed the length of the third, so each piece must have length < 1/2*

___
# Let's label the three pieces $x$, $y$, and $z$

### Then, we know that no piece can exceed the length of the other two combined, i.e.

# $x < y + z$

# $y<x+z$

# $z<x+y$

# We also know that the length of $z$ is simply $1 - (x+y)$ since the sum of the three pieces has length 1. So these conditions become:

# $x < y + (1 - (x+y)) \implies x < \frac{1}{2}$ (1)

# $y<x+(1 - (x+y)) \implies y < \frac{1}{2}$ (2)

# $(1 - (x+y))< (x+y) \implies (x+y) > \frac{1}{2}$ (3)

# So now, we can visualize these conditions on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
x = np.arange(0,1,0.01)
y = 0.5 - x

plt.axhline(0.5, color = 'r')
plt.axvline(0.5, color = 'g')
plt.plot(x, y, color = 'b')
plt.xlim([0,1])
plt.ylim([0,1])
```

## The green line represents condition (1); all $x$ values to the right of the line violate the condition

## The red line represents condition (2); all $y$ values above the line violate the condition

## The blue line represents condition (3); for a given $x$ value, the $y$ value must be above the blue line

# So we can plot the area that satisfies all three conditions

```python
x = np.arange(0,0.5,0.01)
y = 0.5 - x

plt.fill_between(x, 0.5, y, color = 'b')

plt.xlim([0,1])
plt.ylim([0,1])
```

# This is simply a triangle of area $\frac{0.5\cdot0.5}{2} = \frac{1}{8}$

# But we have to consider the set of possible combinations of $x$ and $y$

# We know that if we break the stick into three pieces, the sum of the three must be less than 1

### We can plot this area

```python
x2 = np.arange(0,1,0.01)
y2 = 1 - x2

plt.fill_between(x2, 0, y2, color = 'r')

x1 = np.arange(0,0.5,0.01)
y1 = 0.5 - x1

plt.fill_between(x1, 0.5, y1, color = 'b')

plt.xlim([0,1])
plt.ylim([0,1])
```

# Therefore, $P(\text{Three pieces form a triangle}) = \frac{\frac{1}{8}}{\frac{1}{2}} = \frac{1}{4}$
