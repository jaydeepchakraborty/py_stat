# A die is rolled twice

# Let $X$ denote the sum of the two numbers that turn up

# Let $Y$ be the difference between the two numbers

## i) Show that $E(XY) = E(X)E(Y)$

## ii) Are $X$ and $Y$ independent?

____

# i)

## First, we summarize the probabilities for $X$ and $Y$

```python
import numpy as np
import pandas as pd
```

```python
df_results = pd.DataFrame(columns = ['r1','r2','X','Y'])

i = 0
for r1 in range(1, 7):
    for r2 in range(1, 7):
        X = r1 + r2
        Y = np.abs(r1 - r2)
        df_results.loc[i] = [r1, r2, X, Y]
        i+=1
```

```python
df_results['XY'] = df_results['X']*df_results['Y']
```

```python
df_results
```

- Since our dataframe has 36 rows (one for each **equally likely** roll), we can easily show that $E(XY) = E(X)E(Y)$ by taking the mean of the 'XY' column and comparing it to the mean of column 'X' times the mean of column 'Y' 

```python
df_results['XY'].mean()
```

```python
df_results['X'].mean()*df_results['Y'].mean()
```

# They're the same $\implies E(XY) = E(X)E(Y)$

_____

# ii)

# If $X$ and $Y$ are independent, then $P(X=x|Y=x) = P(X=x)$

## Let's let $Y = 2$ and look at the possible values of $X$

```python
df_results[df_results['Y']==2]
```

# We can see from the table above that when $Y=2$, then $X$ cannot be equal to 2 therefore $P(X=2|Y=2) = 0$, but $P(X=2) = \frac{1}{36} \implies$ not independent
