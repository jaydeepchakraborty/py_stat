# Suppose you toss a dart at a circular target of radius 10 inches

# Given that the dart lands in the upper half, find the probability that:

## a) it lands in the right half of the target

## b) its distance from the center is less than 5 inches

## c) its distance from the center is greater than 5 inches

## d) it lands within 5 inches of the point (0,5)

___

# a)

### First, we visualize the problem

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
x = np.arange(-10,10.1,0.1)
x2 = np.arange(0,10.1,0.1)
y = np.sqrt(100-x**2)
y_below = -np.sqrt(100-x**2)
y2 = np.sqrt(100-x2**2)
```

```python
plt.fill_between(x, y, 0, color = 'r')
plt.fill_between(x, y_below, 0, color = 'g')
plt.fill_between(x2, y2, 0, color = 'b')
```

### The total of area inside the circle (i.e. the green, red and yellow combined) represent the total dartboard

### We know that the dart landed on the upper half, so we don't need to worry about the green lower half

### We can see right away that $P(\text{Dart lands on upper half}) = \frac{1}{2}$

#### We can also see that one quarter of the board is blue, representing $\text{Dart lands on upper half}\cap\text{Dart lands on right half}$

### So $P(\text{Dart lands on upper half}\cap\text{Dart lands on right half}) = \frac{1}{4}$

### Putting it all together:

### $P(\text{Dart lands on right half }|\text{Dart lands on upper half}) = \frac{P(\text{Dart lands on upper half}\cap\text{Dart lands on right half})}{P(\text{ Dart lands on upper half})} = \frac{\frac{1}{4}}{\frac{1}{2}} = \frac{1}{2}$


# We could have also just noticed that half of the upper half lies on the right side of the board, and hence the probability is $\frac{1}{2}$

___

# b)

### First, let's look at the board

```python
x = np.arange(-10,10.1,0.1)
x2 = np.arange(-5,5.1,0.1)
y = np.sqrt(100-x**2)
y2 = np.sqrt(25-x2**2)

y_lower = -y
y2_lower = -y2
```

```python
plt.fill_between(x, y, 0, color = 'r')
plt.fill_between(x, y_below, 0, color = 'g')
plt.fill_between(x2, y2, 0, color = 'b', alpha = 0.5)
plt.fill_between(x2, y2_lower, 0, color = 'b', alpha = 0.5)
```

### So, we can solve for this probability as: $P(\text{Dart lands in purple area | Dart lands on upper half}) = \frac{\text{Purple Area}}{\text{Purple Area + Red Area}}$

### $\text{Purple Area} = \frac{1}{2}\pi(5^{2}) = \frac{25\pi}{2}$ ; $\text{Purple Area + Red Area} = \frac{100\pi}{2}$

## Therefore $P(\text{Dart lands in purple area | Dart lands on upper half}) = \frac{\frac{25\pi}{2}}{\frac{100\pi}{2}} = \frac{1}{4}$

___

# c)

### This is very similar to b), except it's asking for $\frac{\text{Red Area}}{\text{Purple Area + Red Area}} = 1 - \frac{\text{Purple Area}}{\text{Purple Area + Red Area}} = \frac{3}{4}$

# d)

### Again, let's visualize the problem

```python
x_board = np.arange(-10,10.1,0.1)
y1_board = np.sqrt(100-x_board**2)
y2_board = -y1_board

x_circle = np.arange(-5,5.1,0.1)
y1_circle = np.sqrt(25-x_circle**2)+5
y2_circle = -y1_circle+10
```

```python
plt.fill_between(x_board, y1_board, 0, color = 'r')
plt.fill_between(x_board, y2_board, 0, color = 'g')
plt.fill_between(x_circle, y1_circle, 5, color = 'b', alpha = 0.5)
plt.fill_between(x_circle, y2_circle, 5, color = 'b', alpha = 0.5)
```

### So this is asking for the ratio of the purple area to the whole upper half

### The area of the purple circle is equal to $\pi(5^{2}) = 25\pi$

### We know from above that the red area is equal to $50\pi$

## Therefore $\frac{\text{Purple Area}}{\text{Red Area + Purple Area}} = \frac{25\pi}{50\pi} = \frac{1}{2}$
