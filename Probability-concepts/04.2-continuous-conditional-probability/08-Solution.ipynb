# Let $x$ and $y$ be chosen at random from the interval $[0,1]$

# Which pairs of the following are independent:

### a) $x>\frac{1}{3}$

### b) $y>\frac{2}{3}$

### c) $x>y$

### d) $x+y<1$

___

## First off, we know from [Q7](07-Solution.ipynb) that a) and b) are independent

## Let's plot the conditions to see if they make more sense

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,10
```

# Looking at c) and d)

```python
x = np.arange(0,1.01,0.01)
y = x
plt.fill_between(x,y,0,color = 'r', alpha = 0.5)

y = 1-x
plt.fill_between(x,y,0,color = 'y', alpha = 0.5)
plt.xlim([0,1])
```

# We know that each has probability $\frac{1}{2}$, and the intersection of the two triangles is $\frac{1}{2}$ of each, therefore c) and d) are independent

___
# Now looking at c) and a)

```python
x = np.arange(0,1.01,0.01)
y = x
plt.fill_between(x,y,0,color = 'r', alpha = 0.5)
plt.fill_between(np.arange(0.333,1.01,0.01), 0, 1, color = 'b',alpha = 0.5)
plt.xlim([0,1])
```

# Since the purple area isn't exacly half the vertical rectangle, we know c) and a) are not independent

# We can repeat this for the remaining combinations, but it's already clear that (a,b) and (c,d) are the only independent pairs
