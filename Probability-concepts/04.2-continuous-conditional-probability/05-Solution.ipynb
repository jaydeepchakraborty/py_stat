# Suppose you choose two numbers $x$ and $y$ independently at random from the interval $[0,1]$

# Given that their sum lies in the interval $[0,1]$, find the probability that

## a) $|x-y|<1$

## b) $xy<\frac{1}{2}$

## c) $max\{x,y\}<\frac{1}{2}$

## d) $x^{2}+y^{2} < \frac{1}{4}$

## e) $x>y$

___

# a)

# We start off by visualizing the conditions

# Note: for now, we'll ignore the condition that $x+y<1$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 5,10
```

### $|x-y|<1$ means the distance between the x and y-values is always less than one

### Since we know $x$ and $y$ are both in the interval $[0,1]$, we can visualize these conditions as:

```python
x = np.arange(0,1.01,0.01)
y1 = np.ones(len(x))
y2 = x+1
```

```python
plt.fill_between(x, y1, 0, color = 'r')
plt.fill_between(x, y1, y2, color = 'g')
```

### The combination of green and red areas represents the possible combinations of $x$ and $y$ that satisfy the condition that $|x-y|<1$

### Why?

- Let's consider the point 0.4
    - The maximum value in the green area at x=0.4 is just below 1.4, and its minimum value is 0
        - $|0.4-1.399999|<1$ and $|0.4-0|<1$

### But the red area represents the condition that both $x$ and $y$ are in the interval $[0,1]$

### Since this red area is contained in the area that satisfies the condition $|x-y|<1$, we can conclude

# $x, y$ in $[0,1] \implies |x-y|<1 \implies P((|x-y|<1)|(x,y\text{ in }[0,1])) = 1$

___
# b)

### We can visualize $xy<\frac{1}{2}$ and $x+y < 1$ as

```python
x = np.arange(0,1.001,0.001)
y1 = 1-x
y2 = 1/(2*x)
```

```python
plt.rcParams['figure.figsize'] = 10,10
```

```python
plt.fill_between(x, y2, 0, color = 'g')
plt.fill_between(x, y1, 0, color = 'r')
plt.ylim([0,1])
plt.show()
```

### The red triangle represents the possible combinations of $x$ and $y$ that satisfy the condition that $x+y<1$

### The combination of the red and green areas represents the combinations of $x$ and $y$ that satisfy the condition that $xy<\frac{1}{2}$

### But this means that $x+y<1 \implies xy<\frac{1}{2}$ therefore $P((xy<\frac{1}{2})|(x+y<1)) = 1$

___
# c)

### We can visualize $max\{x,y\}<\frac{1}{2}$ and $x+y < 1$ as

```python
x1 = np.arange(0,1.001,0.001)
y1 = 1-x
x2 = np.arange(0,0.501,0.001)
y2 = 0.5*np.ones(len(x2))
```

```python
plt.fill_between(x1, y1, 0, color = 'r')
plt.fill_between(x2, y2, 0, color = 'b')
plt.show()
```

### The blue square represents the possible $x$ and $y$ values that satisfy the condition $max\{x,y\}<\frac{1}{2}$ while the entire red triangle (including the blue square) represents the $x$ and $y$ values that satisfy the condition $x+y < 1$

# So, it's pretty obvious that $P((max\{x,y\}<\frac{1}{2})|(x+y < 1)) = \frac{1}{2}$

___
# d)

### We can visualize $x^{2} + y^{2} < \frac{1}{4}$ and $x+y < 1$ as

```python
x1 = np.arange(0,1.001,0.001)
y1 = 1-x
x2 = np.arange(0,0.501,0.001)
y2 = np.sqrt(0.25-x2**2)
```

```python
plt.fill_between(x1, y1, 0, color = 'r')
plt.fill_between(x2, y2, 0, color = 'b')
```

### The area of the entire red triangle is equal to $Area_{Triangle} = \frac{1}{2}$

### The area of the blue area is equal to $Area_{Blue} = \frac{1}{4}\pi(\frac{1}{2})^{2} = \frac{\pi}{16}$

## Therefore $P((x^{2}+y^{2}<\frac{1}{4})|(x+y<1)) = \frac{\frac{\pi}{16}}{\frac{1}{2}} = \frac{\pi}{8}$

___

# e)

### We can visualize $x>y$ and $x+y < 1$ as

```python
x = np.arange(0,1.001,0.001)
y1 = 1-x
y2 = x
```

```python
plt.fill_between(x, y1, 0, color = 'r')
plt.fill_between(x, y2, 0, color = 'b', alpha = 0.5)
```

### The purple area makes up half the total red triangle

## Therefore $P((x>y)|(x+y < 1)) = \frac{1}{2}$
