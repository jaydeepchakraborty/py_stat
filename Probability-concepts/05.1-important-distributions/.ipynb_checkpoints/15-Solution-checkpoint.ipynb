# Write a program for the user to input $n,p,k$ and have the program print out both 

## 1. $b(n,p,k)$

## 2. The poisson approximation

___

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
from math import factorial
```

```python
def b(n,p,k):
    q = 1-p
    choice = factorial(n)/(factorial(k)*factorial(n-k))
    return choice*(p**k)*(q**(n-k))
```

```python
def poisson(n,p,k):
    lambdA = float(p)*n
    fraction = (lambdA**k)/factorial(k)
    return fraction*(np.exp(1)**(-lambdA))
```

```python
def comparison(n,p,k):
    bin_val = b(n,p,k)
    poiss_val = poisson(n,p,k)
    return bin_val,poiss_val
```

```python
p = 0.5

for n in [10,100]:
    x = range(n+1)
    y1 = []
    y2 = []
    for k in x:
        bin_val,poiss_val = comparison(n,p,k)
        y1.append(bin_val)
        y2.append(poiss_val)
    plt.plot(x,y1, label = 'Binomial')
    plt.plot(x,y2, label = 'Poisson')
    plt.legend()
    plt.title('p = '+str(p)+'; n = '+str(n))
    plt.show()
```
