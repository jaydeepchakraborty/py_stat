# 1. Modify the program CoinTosses to toss a coin n times and print out after every 100 tosses the proportion of heads minus 1/2. Do these numbers appear to approach 0 as $n$ increases? Modify the program again to print out, every 100 times, both the following quantities: the proportion of heads minus 1/2, and the number of heads minus half the number of tosses. Do these numbers appear to approach 0 as $n$ increases?

```python
import numpy as np
```

```python
def simulation(n_iterations):
    list_results = []
    list_proportions = []
    for i in range(n_iterations):
        array = np.random.randint(0,2,size=100)
        list_results = list_results+[x for x in array]
        list_proportions.append(np.mean(list_results)-0.5)
    return list_proportions
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5

N_trials = 1000
plt.plot(range(N_trials), simulation(N_trials))
plt.axhline(0, ls = 'dashed', color = 'r')
```

```python
def simulation2(n_iterations):
    list_results = []
    list_proportions = []
    list_differences = []
    for i in range(n_iterations):
        array = np.random.randint(0,2,size=100)
        list_results = list_results+[x for x in array]
        list_proportions.append(np.mean(list_results)-0.5)
        list_differences.append(np.sum(list_results) - float(len(list_results))/2)
    return list_proportions,list_differences
```

```python
N_trials = 1000
list_proportions,list_differences = simulation2(N_trials)

plt.plot(range(N_trials), list_differences)
plt.axhline(0, ls = 'dashed', color = 'r')
```
