Modify the program CoinTosses so that it tosses a coin $n$ times and records whether or not the proportion of heads is within 0.1 of 0.5 (i.e. between 0.4 and 0.6) Have your program repeat this experiment 100 times. About how large must $n$ be so that approximately 95 times out of 100 the proportion is between 0.4 and 0.6?

```python
import numpy as np
```

```python
def CoinTosses(n):
    results = []
    for i in range(100):
        array = np.random.randint(0,2,size = n)
        proportion = np.mean(array)
        results.append(1*(abs(proportion-0.5)<=0.1))
    return np.sum(results)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
x = range(1,200)
y = [CoinTosses(i) for i in x]
```

```python
plt.plot(x,y)
plt.axhline(95, ls = 'dashed', color = 'r')
```

```python
#adjusting CoinTosses to do 3 trials for each n and take the mean (to smooth out results)
def CoinTosses(n):
    results = []
    for i in range(100):
        list_temp = []
        for j in range(3):
            array = np.random.randint(0,2,size = n)
            proportion = np.mean(array)
            list_temp.append(proportion)
        proportion = np.mean(list_temp)
        results.append(1*(abs(proportion-0.5)<=0.1))
    return np.sum(results)
```

```python
x = range(1,200)
y = [CoinTosses(i) for i in x]
plt.plot(x,y)
plt.axhline(95, ls = 'dashed', color = 'r')
```
