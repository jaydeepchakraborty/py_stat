# Another form of bet for roulette is to bet that a specific number (say 17) will turn up. If the ball stops on your number, you get your dollar back plus an additional 35 dollars. If not, you lose your dollar.

## Write a program that will plot your winnings when you make 500 plays of roulette at Las Vegas, when you bet each time on number 17. How is this chart different from the chart for quesiton 6?

```python
import numpy as np
```

```python
def check(val):
    if val==17:
        return 35
    else:
        return -1
```

```python
check = np.vectorize(check)
```

```python
def spins(n_spins):
    rand_array = np.random.randint(1, 39, size = n_spins)
    return np.cumsum(check(rand_array))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
N_spins = 500
N_trials = 100

for i in range(N_trials):
    x = range(N_spins)
    y = spins(N_spins)
    plt.plot(x,y)
plt.axhline(0, ls = 'dashed', color = 'r')
plt.show()
```

# Seems like more paths have a positive outcome
