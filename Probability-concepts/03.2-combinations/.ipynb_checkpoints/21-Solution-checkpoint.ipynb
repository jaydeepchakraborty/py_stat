# A lady wishes to color her fingernails on one hand using at most two of the colors [red, yellow, blue]

# How many ways can she do this?

___
### If we choose two colors, they can be applied to her fingers $2^{5}$ different ways

### But there are $\binom{3}{2}$ possible ways to choose the two colors

## So, the total number of ways she can do this is $\binom{3}{2}2^{5} = 3\cdot32 = 96$

# NOT DONE YET: we've double counted for some of the possibilities

### Let's say we choose red and yellow

### Then, there are two possible applications of the $2^{5}$ possibilities that are a single color - all red, and all yellow

### But if we choose red and blue, then there are also two single color possibilities - all red and all green

### Similarly, if we choose yellow and blue, then there's all yellow and all blue

## So when we count the 96 possibilities, we're counting the pure color possibilities twice each, and therefore we need to remove 3

# So, the total number of possibilities is equal to 96-3 = 96

___
We can prove this by listing out all possible combinations

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = range(1,6))

colors = ['R','Y','B']

i = 0
for finger_1 in colors:
    for finger_2 in colors:
        for finger_3 in colors:
            for finger_4 in colors:
                for finger_5 in colors:
                    df.loc[i] = [finger_1, finger_2, finger_3, finger_4, finger_5]
                    i+=1
```

```python
df
```

```python
len(df)
```

## Our loop generated 243 rows, which is equal to $3^{5}$, so that's the correct number of combinations

### Now, we count only the rows that have 2 colors or fewer

```python
df2 = df[df.apply(lambda x: len(x.drop_duplicates()), axis = 1)<=2]
df2
```
