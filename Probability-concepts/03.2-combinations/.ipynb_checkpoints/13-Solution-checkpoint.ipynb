# If a set has $2n$ elements, show that it has more subsets with $n$ elements than with any other number of elements

___

### We can show this by picking some value of 2n, and looping through all possible sizes of subsets to see how many possibilities there are

### If we divide the number of possibilities by $\binom{2n}{n}$, we'll see if any are larger

```python
from math import factorial
import numpy as np
```

```python
def choose(n, k):
    numerator = factorial(n)
    denominator = factorial(k)*factorial(n-k)
    return numerator/float(denominator)
```

```python
def comparing_size(n,k):
    numerator = choose(n, k)
    denominator = choose(n, n/2)
    return numerator/denominator
```

```python
comparing_size = np.vectorize(comparing_size)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
for n in [10,20,50,100]:
    x = np.arange(n+1)
    y = comparing_size(n, x)
    plt.plot(x,y)
    plt.axvline(n/2, ls = 'dashed', color = 'r')
    plt.title(str(n))
    plt.show()
```

# All are symmetrical around a peak at the midpoint

___

# Alternative solution

### We know that if we have a set $S$ containing $2n$ elements, the number of subsets with $k$ elements is given by 

### $\binom{2n}{k} = \frac{2n!}{k!(2n-k)!}$

### So as we try different $k$ values, only the denominator changes

### Therefore, to prove that $\binom{2n}{n}$ has the highest number of combinations, we just need to show that $k!(2n-k)!$ is smallest for $k=n$

```python
def denominator_comparison(n, k):
    return factorial(k)*float(factorial(n-k))
```

```python
denominator_comparison = np.vectorize(denominator_comparison)
```

```python
for n in [10,20,50,100]:
    x = np.arange(n+1)
    #note: plotting log scale to shrink y axis
    y = np.log(denominator_comparison(n, x))
    plt.plot(x,y)
    plt.axvline(n/2, ls = 'dashed', color = 'r')
    plt.title(str(n))
    plt.show()
```
