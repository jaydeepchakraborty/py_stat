# Suppose that $X$ and $Y$ are independent and $Z=X+Y$. Find $f_{Z}$ if

## a) $f_{X}(x) = \lambda e^{-\lambda x} \text{ if }x>0$ and $f_{Y}(y) = \mu e^{-\mu Y} \text{ if }y>0$

## b) $f_{X}(x) = \lambda e^{-\lambda x} \text{ if }x>0$ and $f_{Y}(y) = 1 \text{ if }0\leq y \leq 1$

_____

# a)

# Just to see what the distribution looks like, let's set $\lambda = 1$ and $\mu = 2$ and take some samples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

- To take samples from $X$, we know solve for $x$ such that $rand = 1 -  e^{-x} \implies x = ln(\frac{1}{1-rand})$
- To take samples from $Y$, we know solve for $y$ such that $rand = 1-e^{-2y} \implies x = \frac{1}{2}ln(\frac{1}{1-rand})$

```python
n = 100000
x = np.random.random(size = n)
y = np.random.random(size = n)

X = np.log(1/(1-x))
Y = 0.5*np.log(1/(1-y))

Z = X + Y

plt.hist(Z, bins = 100, normed = True)
plt.xlim([0,10])
plt.show()
```


# $f_{Z}(z) = \int_{-\infty}^{\infty}f_{X}(x)f_{Y}(z-x)dx$

# When $x\leq 0, f_{X}(x) = 0 \implies f_{Z}(z) = \int_{0}^{\infty}\lambda e^{-\lambda x}f_{Y}(z-x)dx$

# We also know that when $z-x \leq 0$,  $f_{Y}(z-x) = 0 $

# $\implies f_{Z}(z) = \int_{0}^{z}\lambda e^{-\lambda x}\mu e^{-\mu (z-x)}dx = -\frac{\lambda \mu e^{-\lambda z} - e^{-\mu z}}{\lambda - \mu} = \frac{\lambda \mu e^{-\lambda z} - e^{-\mu z}}{\mu-\lambda}$

# So when $\lambda = 1$ and $\mu = 2$ this becomes $2(e^{-z} - e^{-2z})$

```python
z = np.arange(0, 10, 0.01)
fz = 2*(np.exp(-z)-np.exp(-2*z))
plt.plot(z, fz, color = 'r', lw = 3)
plt.hist(Z, bins = 100, normed = True)
plt.xlim([0,10])
plt.show()
```

# b)

## Let's start by plotting the density when $\lambda = 1$

- To take samples from $X$, we know solve for $x$ such that $rand = 1 -  e^{-x} \implies x = ln(\frac{1}{1-rand})$
- To take samples from $Y$, we know solve for $y$ such that $rand = y \implies y = rand$

```python
n = 100000
x = np.random.random(size = n)

X = np.log(1/(1-x))
Y = np.random.random(size = n)

Z = X + Y

plt.hist(Z, bins = 100, normed = True)
plt.xlim([0,10])
plt.show()
```


# $f_{Z}(z) = \int_{-\infty}^{\infty}f_{X}(x)f_{Y}(z-x)dx$

# When $x\leq 0, f_{X}(x) = 0 \implies f_{Z}(z) = \int_{0}^{\infty}\lambda e^{-\lambda x}f_{Y}(z-x)dx$

# We also know that we need $0\leq z-x \leq 1$ for $f_{Y}(z-x)$ to be non-zero

# This implies that we need to integrate over $z-1 \leq x \leq z$

# If $0\leq z \leq 1$ then $z-1 \leq 1 \implies f_{Z}(z) = \int_{0}^{z}\lambda e^{-\lambda x}dx = 1 - e^{-\lambda z}$

# If $1\leq z$ then we have an upper bound of 1 $\implies f_{Z}(z) = \int_{z-1}^{1}\lambda e^{-\lambda x}dx = e^{-\lambda (z-1)} - e^{-\lambda z}$

# Therefore

# $f_{Z}(z) = \left\{\begin{matrix}1-e^{-\lambda z} & 0\leq z \leq 1\\ e^{-\lambda(z-1)} - e^{-\lambda z} & 1 \leq z\end{matrix}\right.$

```python
z1 = np.arange(0, 1, 0.01)
fz1 = 1 - np.exp(-z1)

z2 = np.arange(1, 10, 0.01)
fz2 = np.exp(-1*(z2-1))-np.exp(-z2)

plt.plot(z1, fz1, color = 'r', lw = 3)
plt.plot(z2, fz2, color = 'r', lw = 3)


plt.hist(Z, bins = 100, normed = True)
plt.xlim([0,10])
plt.show()
```
