# Suppose that $R^{2} = X^{2} + Y^{2}$. Find $f_{R^{2}}$ and $f_{R}$ if

# $f_{X}(x) = f_{Y}(x) = \frac{1}{2}$ when $-1\leq x \leq 1$

______

# First, let's think about what $X^{2}$ looks like

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
n = 10000
X = np.random.uniform(-1, 1, size = n)
X2 = X**2
plt.hist(X2, bins = 100, normed = True)
plt.show()
```

# Since $X$ is uniformly distributed, we know 

# $P(X\leq x) = \int_{-1}^{x}\frac{1}{2}dx = \frac{x}{2}|^{x}_{-1} = \frac{x+1}{2}$

# $\implies P(X^{2}\leq x) = P(|X|\leq \sqrt{x}) = 2\frac{\sqrt{x} + 1}{2} \implies f_{X^{2}}(x) = \frac{1}{2\sqrt{x}}$

```python
x2 = np.arange(0.01,1,0.01)
fx2 = 1/(2*np.sqrt(x2))

plt.plot(x2, fx2, color = 'r', lw = 3)
plt.hist(X2, bins = 100, normed = True)
plt.show()
```

# By the same logic, we know that $f_{Y^{2}}(y) = \frac{1}{2\sqrt{y}}$

# Combining these, we know $f_{R^{2}}(r) = \int_{-\infty}^{\infty}f_{X^{2}}(x)f_{Y^{2}}(r-x)dx$


# We know $0\leq x \leq 1$ and $0 \leq r-x \leq 1$ for $f_{X^{2}}(x)$ and $f_{Y^{2}}(r-x)$

# But $0\leq r \leq 2$


____

# If $0\leq r \leq 1$ then $r-1 \leq 0 \implies f_{R^{2}}(r) = \int_{0}^{r}\frac{1}{2\sqrt{x}}\frac{1}{2\sqrt{r-x}}dx = ... = \frac{\pi}{4}$

____

# If $1\leq r\leq 2 \implies f_{R^{2}}(r) = \int_{r-1}^{2}\frac{1}{2\sqrt{x}}\frac{1}{2\sqrt{r-x}}dx = ... $

# $= 2 \left ( sin^{-1}(\frac{\sqrt{2}}{\sqrt{r}}) - tan^{-1}(\sqrt{r-1}) \right )$

_____

# Let's see if this fits

```python
n = 100000
X = np.random.uniform(-1, 1, size = n)
X2 = X**2

Y = np.random.uniform(-1, 1, size = n)
Y2 = Y**2

R2 = X2 + Y2

r1 = np.arange(0, 1, 0.01)
fr1 = np.pi/4*np.ones(len(r1))

r2 = np.arange(1,2, 0.01)
fr2 = 0.5*np.arcsin((2-r2)/r2)

plt.plot(r1, fr1, color = 'r', lw = 3)
plt.plot(r2, fr2, color = 'r', lw = 3)

plt.hist(R2, bins = 100, normed = True)
plt.show()
```
