# Let $X$, $Y$, and $Z$ be independent random variables with:

# $ f_{X}(x) = f_{Y}(y) = f_{Z}(x) = \left\{\begin{matrix}1 & 0 \leq x \leq 1\\ 0 & otherwise\end{matrix}\right.$

# Suppose that $W = X + Y + Z$. 

# Find $f_{W}$ directly, and compare your answer to the one given by the formula in Example 7.9

_____

# $X$, $Y$, and $Z$ all have the standard uniform density

## Let's take samples from each and plot them

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
n = 100000
X = np.random.random(size = n)
Y = np.random.random(size = n)
Z = np.random.random(size = n)

W = X+Y+Z

plt.hist(W, bins = 100, normed = True)
plt.show()
```

# Let $\bar{X} = X+Y \implies f_{\bar{X}}(\bar{x}) = \left\{\begin{matrix}\bar{x} & 0 \leq \bar{x} \leq 1\\ 2-\bar{x} & 1 \leq \bar{x} \leq 2\end{matrix}\right.$

# Then $W = X + Y + Z = \bar{X} + Z$

# Now, we have $f_{W}(w) = \int_{-\infty}^{\infty}f_{\bar{X}}(\bar{x})f_{Z}(w-\bar{x})d\bar{x}$

# We know that if $w-\bar{x} < 0$ or $1 < w-\bar{x}$ then $f_{Z} = 0$, otherwise it's 1 (so we can replace it with 1)

# $\implies f_{W}(w) = \int_{w-1}^{w}f_{\bar{X}}(\bar{x})d\bar{x}$

# If $0\leq w \leq 1$, then $w-1 \leq 0$

# $\implies \int_{w-1}^{w}f_{\bar{X}}(\bar{x})d\bar{x} = \int_{w-1}^{0}f_{\bar{X}}(\bar{x})d\bar{x} + \int_{0}^{w}f_{\bar{X}}(\bar{x})d\bar{x} = 0 + \int_{0}^{w}f_{\bar{X}}(\bar{x})d\bar{x}$

# Since $w \leq 1$, we know $f_{\bar{X}}(\bar{X}) = \bar{x} \implies f_{W}(w) = \frac{w^{2}}{2}$

___

# Next, if $1\leq w \leq 2$, we know $0 \leq w-1 \leq 1$

# $\implies \int_{w-1}^{w}f_{\bar{X}}(\bar{x})d\bar{x} = \int_{w-1}^{1}\bar{x}d\bar{x} + \int_{1}^{w}(2-\bar{x})d\bar{x} = 3w - w^{2} - \frac{3}{2}$

_____

# Finally, if $2 \leq w \leq 3$, then $f_{\bar{X}}(w) = 0$

# $\implies \int_{w-1}^{w}f_{\bar{X}}(\bar{x})d\bar{x} = \int_{w-1}^{2}(2-\bar{x})d\bar{x} = \frac{(w-3)^{2}}{2}$

# Therefore

# $f_{W}(w) = \left\{\begin{matrix}\frac{w^{2}}{2} & 0\leq w \leq 1\\ 3w - w^{2} - \frac{3}{2} & 1\leq w \leq 2\\ \frac{(w-3)^{2}}{2} & 2\leq w \leq 3\end{matrix}\right.$

```python
w1 = np.arange(0, 1, 0.01)
w2 = np.arange(1, 2, 0.01)
w3 = np.arange(2, 3, 0.01)

fw1 = w1**2/2
fw2 = 3*w2 - w2**2 - 1.5
fw3 = (w3 - 3)**2/2

plt.plot(w1, fw1, color = 'r', lw = 3)
plt.plot(w2, fw2, color = 'r', lw = 3)
plt.plot(w3, fw3, color = 'r', lw = 3)
plt.hist(W, bins = 100, normed = True)
plt.show()
```
