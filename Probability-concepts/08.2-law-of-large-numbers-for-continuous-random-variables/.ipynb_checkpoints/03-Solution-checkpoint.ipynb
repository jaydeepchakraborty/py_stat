# Let $X$ be the random variable of [Exercise 2](02-Solution.ipynb)

## a) Calculate the function $f(x) = P(\left | X-10 \right |\geq x)$

## b) Now graph the function $f(x)$ and on the same axes graph the Chebyshev function $g(x) = \frac{100}{3x^{2}}$. Show that $f(x)\leq g(x)$ for all $x>0$ but that $g(x)$ is not a very good approximation

_______

# a)

# $f(x) = P(X\leq 10 - x) + \left (1 - P(X < 10+x) \right ) = \frac{10-x}{20} + (1-\frac{10+x}{20})$

# $= 1-\frac{x}{10}$

_____

# b)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
x = np.arange(11)
f = 1 - x/float(10)
g = float(100)/(3*(x+0.0000000000000001)**2)
```

```python
plt.plot(x,f)
plt.plot(x,g)
plt.ylim(0,1)
```
