# The LLN says that $A_{n} \rightarrow E(X_{i})$ as $n\rightarrow \infty$

## What about $\frac{Z_{n}}{m^{n}}$? Does that converge to some constant?

## We'll see that unlike the LLN, $\frac{Z_{n}}{m^{n}}$ converges *not to a constant, but to a random variable*

### Instead of proving it, we'll use a simulation to demonstrate

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
array = [0.2092, 0.2584, 0.2360, 0.1593, 0.0828, 0.0357, 0.0133, 0.0042, 0.0011]
cumsum_array = np.cumsum(array)
```

```python
def sim_single_family(n):
    rand_array = np.random.random(size = n)
    return np.searchsorted(cumsum_array,rand_array)
```

```python
def sim(n):
    Z = 1
    y = [Z]
    
    for i in range(1,n+1):
        Z_new = np.sum(sim_single_family(Z))
        Z = Z_new
        y.append(Z)
    
    return y
```

```python
x = np.arange(26)

for i in range(10):
    y = sim(25)
    plt.plot(x,y)
    plt.title('Growth of process')
```

```python
mn_array = 1.837**(x)
```

```python
for i in range(10):
    y = sim(25)/mn_array
    plt.plot(x,y)
    plt.title('Growth of Zn/m^n')
```

## As we can see, each process converges to some stable value, but that stable value is a random variable

# *Chain Letter Problem*

### In 1978, a chain letter called the "Circle of Gold" started in California

### The chain required a participant to buy a letter containing a list of 12 names for 100 dollars

### The buyer gives 50 dollars to the person who gave them the letter, and sends 50 dollars to the first of the 12 names

### The buyer then crosses the top name off the list and adds theirs to the bottom of the list

### So if I buy the letter for 100 dollars, I can sell it and get 50 back, and I get another 50 back if the letter is sold another 12 times and I get 50 dollars in the mail

## Therefore, if the probability of selling the letter is $p$, my expected payoff is $-100 + 50p + 50p^{12}$

### As we can see, this is a pretty unfavourable game

## What if we were allowed to send the letter to 2 people?

### This becomes a branching process

### We assume that every letter is either sold to 0, 1, or 2 people with probabilities $p_{0}$, $p_{1}$, and $p_{2}$

### Let $Z_{1}$, $Z_{2}$,..., $Z_{12}$ be the number of letters in the first 12 generations of the process

## Then your expected payoff is 

# $-100 + 50\cdot E(Z_{1}) + 50\cdot E(Z_{12}) = -100 + 50m + 50m^{12}$

## where $m=p_{1}+2p_{2}$ i.e the expected number of letters each person sells

## So, for our expected payoff to be positive, we need $-100 + 50m + 50m^{12} > 0$

## $\implies m + m^{12} > 2 \implies m>1$

## Recall: we saw that the roots of $d = p_{0} + p_{1}d + p_{2}d^{2}$ are $d=1$ and $d=\frac{p_{0}}{p_{2}}$

## $\implies$ we need $p_{2}>p_{0}$ to have a positive expected payoff

## Let $p_{0} = 0.2$, $p_{1} = 0.5$, and $p_{2} = 0.3 \implies m = 1.1 $

## $\implies E(\text{Payoff}) = -100+50(1.1 + 1.1^{12}) = 111.92$

```python
array = [0.2,0.5,0.3]
cumsum_array = np.cumsum(array)
```

```python
def single_sim(n):
    rand_array = np.random.random(size=n)
    n_letters = np.sum(np.searchsorted(cumsum_array, rand_array))
    return n_letters
```

```python
def sim():
    Z = 1
    y = []
    for i in range(12):
        Z_new = single_sim(Z)
        Z = Z_new
        y.append(Z)
    return -100+50*y[0]+50*y[-1]
```

```python
total = 0
for i in range(200000):
    total += sim()
total/200000.0
```
