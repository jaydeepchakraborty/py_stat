# The Polya urn model for contagion is as follows:

## 0. We start with an urn which contains one white ball and one black ball

## 1. Choose a ball from the urn

## 2. Put the chosen ball back, and add an additional ball to the urn of the same colour

## 3. Go back to step 1

# Write a simulation for this model, and see if you can make any predictions about the proportion of white balls in the urn after a large number of draws

# Is there a tendency to have a large proportion of balls the same color in the long run?

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (15,15)
```

```python
def sim(n_steps):
    urn = [0,1]
    y = []
    for i in range(n_steps):
        choice = np.random.choice(urn)
        urn.append(choice)
        y.append(np.sum(urn)/(i+3))
        
    return y
```

```python
n_steps = 100
n_trials = 10

x = np.arange(n_steps)
for i in range(n_trials):
    y = sim(n_steps)
    plt.plot(x,y)
plt.show()
```
